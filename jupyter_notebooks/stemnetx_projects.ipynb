## Problem 0: Marker in DNA
#### Coders: Troy Coffey, Mariam Owusu, Richmond Boahene, Albany Zavarce

Here is a cool problem having to do with DNA structure.  (Please keep in mind that this problem is written by a particle physicist, not someone with an excellent understanding of DNA.)

You may know that DNA kind of looks like a ladder, with the rungs represented by *base pairs* made of pairs (duh) of *nucleobases*.
Nucleobases are complex molicules that come in four flavors: cytosine [C], guanine [G], adenine [A] or thymine [T].  A and T pair, G and C pair.
Genetic information is encoded in the order/sequence of these base pairs.

In the `data_files` directory, you'll find the following three files:
- dna_1_1000000bps.txt 
- dna_2_1000000bps.txt 
- dna_marker_50bps.txt

The first two files indicate the type of DNA (1 or 2) and the number of base pairs in the file (1000000).
The third file contains a small (50 base pairs) snippet of DNA (marker) that is an indicator of a certain hereditary trait.

#### Goal

Determine whether the dna_1 and dna_2 files contain the marker given in the marker file and, if so, where in each dna sequence the marker lives (*i.e.*, what line of the dna file does the marker sequence begin on?).

#### Steps
1. Write code that reads in a DNA file and the marker file.  Store the base pair info in some type of *container*!
2. Add code that scans through the sequence of base pairs to check whether chunks of 50 base pairs match the marker.
3. Print out the line number/index at which the marker begins in the sequence, and count the number of matches in each file.

## Problem 1: Codon frequency in DNA
#### Coders: Brielle Marnich, Kaylee Huber, Jacob Pittman

Here is another cool problem having to do with DNA structure.  (Please keep in mind that this problem is written by a particle physicist, not someone with a good understanding of DNA structure.)

You may know that DNA kind of looks like a ladder, with the rungs represented by *base pairs* made of pairs (duh) of *nucleobases*.
Nucleobases are complex molicules that come in four flavors: cytosine [C], guanine [G], adenine [A] or thymine [T].  A and T pair, G and C pair.
Genetic information is encoded in the order/sequence of these base pairs.  

Base pairs are grouped into sets of three, called *codons*.  
For example, the following are possible codons: AT-AT-GC, TA-CG-GC.

In this project, you'll determine what is the most common codon in each of the following dna files:
- dna_1_1000000bps.txt 
- dna_2_1000000bps.txt 

The first two files indicate the type of DNA (1 or 2) and the number of base pairs in the file (1000000).

#### Goal

Determine the frequencies (or fractions) of the 8 different codons present in each dna file.
#### Steps
1. Begin by writing some code that reads in a DNA file and makes bar graphs of the frequencies of the four **base pair** types (*i.e.*, NOT nucleobase types).  Develop/test this file on the smaller files before running on the larger files.
2. Write code that reads in a DNA file and the marker file and scans through the sequence of base pairs to check whether chunks of 50 base pairs match the marker.  Find the line number/index at which the marker begins in the sequence and count the number of matches in each file.

## Problem 2: Bacteria growth
#### Coders: Camille Jacome, Madeline Hernandez-Mendez, Samiya Henry, Oyelekan Oyelola

A forensic pathologist studies a tissue sample containing a bacteria culture that was recovered from a body.
The particular bacteria is not known, but it is suspected as the cause of death.
The pathologist needs to determine how fast the bacteria multiply and how many were present in the tissue at the time of death.
Unfortunately, the tissue sample is 24 hours old and no measurements have yet been made.
The pathologist measures the number of bacteria each hour for 6 hours (beginning at 24 hours), and provides the data in the file `bacteria_culture_2025.csv`.

Under ideal conditions a culture grows according to the equation
\begin{equation}
N(t) = N_0 \exp\left(\lambda t \right)
\end{equation}

where
- $N$ is the number of bacteria at time $t$
- $N_0$ is the number of bacteria at time $t = 0$
- $\lambda$ is the growth constant

#### Goal
Fit this function to the pathologist's data to determine $\lambda$ and $N_0$ and their associated uncertainties.

#### Steps
1. Read in the data.
2. Define a function.
3. Use `scipy.optimize` to fir the function to the data.
4. Extract the optimal values of the two fit parameters.
5. Make a graph of the data and the optimized model that includes the time range from $t = 0$ to $t = 31$ hours.
6. Access the elements of the covariance matrix to determine the uncertainty on each fit parameter.
